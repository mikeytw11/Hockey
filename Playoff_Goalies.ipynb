```python
#Import Dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Import Data of Goaltenders during the 2022-2023 Stanley Cup Playoffs
goalies_df = pd.read_csv("https://raw.githubusercontent.com/mikeytw11/Hockey/main/Playoff_Goalies_22_23/EH_std_gl_stats_all_playoffs_adj_2023-05-22.csv")
goalies_df
```

```python
#Create a bar plot using GSAx and Player data from Goalie dataframe.
plt.figure(figsize=(9,6)) 
plt.barh(y=goalies_df['Player'], 
        width=goalies_df['GSAx'], 
        color='green') 
plt.xlabel('GSAx')
plt.ylabel('Player')
plt.title('GSAx Leaderboard, 2022-23 NHL Stanley Cup Playoffs')
plt.show()
```
